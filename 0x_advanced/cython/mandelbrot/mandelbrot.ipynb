## Import c/c++ code

```python
%matplotlib inline
import mandelbrot
import numpy as np
import matplotlib.pyplot as plt
```

## Setup

```python
# Problem size
n = 2048
print(n)
maxtime = 50

# Domain
x = np.linspace(-2.2, 1.5, n)
y = np.linspace(-1.5j, 1.5j, n, dtype='complex64')
x, y = np.meshgrid(x, y)

# Define complex plane
p = x + y 
```

```python
%%time
t = mandelbrot.escape_time(p, maxtime)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(t, cmap='viridis')
```
